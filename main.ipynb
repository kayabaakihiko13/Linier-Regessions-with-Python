# Import Pakage

```python
%matplotlib widget
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.animation import FuncAnimation as anime
```

## Generate Data

```python
data=int(input('masukan jumlah data'))
x_data=[value*0.1 for value in range(data)]
y_data=[value*0.1+np.random.randn() for value in range(data)]
```

# linear gradien

```python
def gradien_linier(x,m):
    return x*m
```

```python
x_prediction=np.array([0,10])
m=5
y_prediction=np.array(gradien_linier(x_prediction,m))
```

```python
plt.scatter(x_data,y_data,c='#031CEE')
plt.plot(x_prediction,y_prediction,c='#000')
plt.axis([0,10,0,10])
plt.show()
```

# Modeling liniear

```python
m_prediction=m
x_prediction_list=[]
y_prediction_list=[]
m_prediction_list=[]
for a in range(data):
    y_prediction=np.array(gradien_linier(x_data[a],m_prediction))
    y_actual=np.array(y_data[a])
    error_data=y_actual-y_prediction
    difference_m=error_data/x_data[a]
    m_prediction=m_prediction+difference_m
    m_prediction_list.append(m_prediction)
    x_prediction_list.append(x_prediction)
    y_prediction_list.append(y_prediction)
```

# Visualisasi Prediksi

```python
fig=plt.figure(figsize=(4,4))
line,=plt.plot([],[],'#E96C03')
def animation(data_data):
    x= x_prediction_list[data_data]
    y=y_prediction_list[data_data]
    line.set_data(x,y)
    return line
plt.scatter(x_data,y_data,c='#031CEE')
plt.axis([0,10,0,10])
prediksi_animasi=anime(fig,animation,frames=100,interval=100,repeat=False)
plt.show()
```
